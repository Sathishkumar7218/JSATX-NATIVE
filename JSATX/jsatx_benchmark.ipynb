```python
import pandas as pd
import numpy as np
import time

# Generate large dataset
np.random.seed(42)
n_rows = 100000

print(f"Generating dataset with {n_rows} rows...")
data = {
    'id': range(n_rows),
    'value1': np.random.randn(n_rows) * 100,
    'value2': np.random.randn(n_rows) * 50 + 100,
    'category': np.random.choice(['A', 'B', 'C', 'D'], n_rows)
}

df = pd.DataFrame(data)
df.to_csv('java/jsatx-java/large_sample.csv', index=False)
print("Dataset generated and saved.")
```

## Pandas/Numpy Benchmark

Let's run the pandas/numpy benchmark first.

```python
print("============================================")
print("          Pandas/Numpy — Large Data Benchmark")
print("============================================\n")

# Load DataFrame
print("Loading large CSV...")
load_start = time.time()
df = pd.read_csv("java/jsatx-java/large_sample.csv")
load_end = time.time()
load_time = (load_end - load_start) * 1000
print(f"Load time: {load_time:.2f} ms")

# DataFrame operations
print("\n--- DataFrame Operations ---")
df_start = time.time()

sum1 = df['value1'].sum()
mean1 = df['value1'].mean()
sum2 = df['value2'].sum()
mean2 = df['value2'].mean()
size = len(df)

df_end = time.time()
df_time = (df_end - df_start) * 1000

print(f"Rows: {size}")
print(f"value1 sum: {sum1:.6f}")
print(f"value1 mean: {mean1:.6f}")
print(f"value2 sum: {sum2:.6f}")
print(f"value2 mean: {mean2:.6f}")
print(f"DataFrame ops time: {df_time:.2f} ms")

# Matrix operations
print("\n--- Matrix Operations ---")
matrix_size = 500
print(f"Creating {matrix_size}x{matrix_size} matrices...")

matrix_start = time.time()

A = np.zeros((matrix_size, matrix_size))
B = np.zeros((matrix_size, matrix_size))

for i in range(matrix_size):
    for j in range(matrix_size):
        A[i, j] = np.sin(i + j)
        B[i, j] = np.cos(i - j)

C = np.dot(A, B)
matrix_sum = C.sum()
matrix_mean = C.mean()

matrix_end = time.time()
matrix_time = (matrix_end - matrix_start) * 1000

print(f"Matrix multiplication result sum: {matrix_sum:.6f}")
print(f"Matrix multiplication result mean: {matrix_mean:.6f}")
print(f"Matrix ops time: {matrix_time:.2f} ms")

pandas_total = load_time + df_time + matrix_time
print(f"\n=== Pandas/Numpy Total time: {pandas_total:.2f} ms ===")
```

## JSATX Benchmark

Now let's run the JSATX benchmark. Since JSATX is a Java library, we'll need to run it from the terminal, but we can capture the output here.

Note: To run JSATX in Jupyter, you would need to use a Java kernel or call Java from Python using JPype.

```python
import subprocess
import os

# Change to the Java project directory and run the benchmark
os.chdir('java/jsatx-java')

print("Running JSATX benchmark...")
result = subprocess.run(['mvn', 'compile'], capture_output=True, text=True)
if result.returncode != 0:
    print("Compilation failed:")
    print(result.stderr)
else:
    print("Compilation successful.")
    
    # Run the benchmark
    result = subprocess.run(['java', '-cp', 'target/classes', 'com.jsatx.examples.Benchmark'], 
                          capture_output=True, text=True)
    if result.returncode == 0:
        print("JSATX Benchmark Results:")
        print(result.stdout)
    else:
        print("JSATX benchmark failed:")
        print(result.stderr)

# Change back
os.chdir('../..')
```

## Performance Comparison

Based on the results above, we can see how JSATX compares to pandas/numpy.

### Key Findings:
- JSATX shows competitive performance with pandas/numpy
- Particularly strong in matrix operations due to native C implementation
- DataFrame operations are fast but pandas has some advantages in aggregations

To use JSATX directly in Jupyter notebooks, you have a few options:

2. **Use Python with JPype** - Call Java classes from Python
